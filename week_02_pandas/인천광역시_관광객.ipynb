## 1. 인천광역시 내 군,구별 핫플레이스

```python
import pandas as pd
import numpy as np
```

```python
# 데이터는 관광지식정보시스템 - 주요관광지점 입장객 통계를 활용하였습니다.
filename ="주요관광지점 입장객(2004.07 이후)_230713041308.xlsx"
```

```python
# 파일읽기
df=pd.read_excel(filename, skiprows=[0],index_col=[3])
```

```python
# 컬럼별 정보 확인, 결측치 확인
df.info()
```

```python
# 실제 데이터 확인
df.head()
```

```python
# 실제 데이터 확인
df.tail()
```

```python
# 결측치 0으로 채움
df= df.fillna(0)
```

```python
# 컬럼명 변경
df=df.rename(columns={"Unnamed: 0":"시도","Unnamed: 1":"군구","Unnamed: 2":"관광지","Unnamed: 4":"총계"})
```

```python
# 인덱스 확인
df.index
```

```python
# 인덱스 내 내국인, 외국인 제거
df_sum_only=df.drop(["내국인","외국인"]) 
```

```python
# 인덱스 리셋
df_sum_only=df_sum_only.reset_index()
```

```python
# 인덱스 새로 지정
df_sum_only=df_sum_only.set_index("군구")
```

```python
df_sum_only.index
```

```python
# 중복없는 인덱스 추출
df_sum_only.index.unique()
```

```python
gu_index=df_sum_only.index.unique()
```

```python
# 딕셔너리에 인천광역시 군,구 저장
dicts={}

for i in gu_index:
    dicts["인천광역시_"+i]= df_sum_only.sort_values(by="총계", ascending=False)
    dicts["인천광역시_"+i]=dicts["인천광역시_"+i].loc[i,:"총계"].drop(columns=["index"])
```

```python
dicts['인천광역시_계양구'] = dicts['인천광역시_계양구'].to_frame()
dicts['인천광역시_계양구']=dicts['인천광역시_계양구'].T
```

```python
dicts['인천광역시_계양구']
```

```python
dicts['인천광역시_계양구']=dicts['인천광역시_계양구'].drop(columns=["index"])
```

```python
dicts['인천광역시_미추홀구'] = dicts['인천광역시_미추홀구'].to_frame()
dicts['인천광역시_미추홀구']=dicts['인천광역시_미추홀구'].T
```

```python
dicts['인천광역시_미추홀구']
```

```python
dicts['인천광역시_미추홀구']=dicts['인천광역시_미추홀구'].drop(columns=["index"])
```

```python
bin_name=["하플","중플","핫플"]

for i in dicts.keys():
        count, bin_dividers = np.histogram(dicts[i]["총계"],bins=3)
        dicts[i]["핫플여부"]=pd.cut(x=dicts[i]["총계"],bins=bin_dividers,labels=bin_name,include_lowest=True)
```

```python
writer = pd.ExcelWriter('인천_핫플.xlsx')
```

```python
for i in dicts.keys():
    dicts[i].to_excel(writer,sheet_name=i)
```

## 2. 인천광역시에서 어떤 군, 구가 제일 핫플?

```python
dict_region= {}
for idx in gu_index:
    dict_region[idx]=df_sum_only.loc[idx,"총계"].sum()
```

```python
# 딕셔너리 -> 시리즈
sr1= pd.Series(dict_region)
```

```python
# 오름차순 정렬
sr1=sr1.sort_values(ascending=False)
sr1 = pd.DataFrame(sr1)
```

```python
sr1.columns=["총방문자수"]
```

```python
sr1
```

```python
bin_name=["하플","중플","핫플"]
count, bin_dividers = np.histogram(sr1["총방문자수"],bins=3)
sr1["핫플여부"]=pd.cut(sr1["총방문자수"],bins=bin_dividers,labels=bin_name,include_lowest=True)
```

```python
sr1.to_csv("핫플.csv",encoding="cp949")
```

## 4. 가장 많이 놀러다니는 월은?

```python
df.columns.unique()
```

```python
ndf=df.drop(columns=["총계","인원계","인원계.1","인원계.2","인원계.3","인원계.4","인원계.5"])
```

```python
ndf=pd.DataFrame(ndf.sum(numeric_only=True))
```

```python
month_dict ={}
for i in range(12):
    month_dict[str(i+1)+"월"]=ndf.iloc[i::12,::].sum()
```

```python
month_df= pd.DataFrame(month_dict)
```

```python
month_df=month_df.T
```

```python
month_df.columns=["6년간 방문자수"]
```

```python
month_df
```

```python
bin_name=["하월","중월","핫월"]
count, bin_dividers = np.histogram(month_df["6년간 방문자수"],bins=3)
month_df["핫월여부"]=pd.cut(month_df["6년간 방문자수"],bins=bin_dividers,labels=bin_name,include_lowest=True)
```

```python
month_df
```

```python
month_df=month_df.rename(columns={"6년간 방문자수":"Number of visitors for 6 years"})
```

```python
month_df.index= ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
```
