```python
#모듈 로딩
import cv2
import os,glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from rembg import remove
```

```python
# 이미지 데이터 관련 변수
path = "./image/dino/"
filename = path + "*m*"
```

```python
# 전체 파일 리스트에 담기
fileList = glob.glob(filename)
```

```python
# 동일 사이즈 조절 후 이미지 데이터 읽기
dsize_ =(100,100)
imgdatas=[]

# 리사이즈, 배경제거
for file in fileList:
    try:
        org =cv2.imread(file,cv2.IMREAD_GRAYSCALE)
        img=cv2.resize(org,dsize_)
        img=remove(img)
        imgdatas.append(img)
    except Exception as e:
        print(str(e))

len(imgdatas)
```

```python
plt.imshow(imgdatas[1].reshape(100,100,4))
```

```python
# 이미지 회전을 통해 이미지 생성
list1= []
for i in imgdatas:
    for degree in range(0,360,10):
        matrix= cv2.getRotationMatrix2D((50,50),degree,1)
        dst =cv2.warpAffine(i,matrix,(100,100))
        dst= dst/255
        dst=dst.reshape(-1)
        list1.append(dst)
```

```python
len(list1)
```

```python
for i in range(len(list1)):
    list1[i]=np.append(list1[i],3)
```

```python
# 이미지 데이터 관련 변수
path = "./image/rabbit/"
filename = path + "*m*"
```

```python
# 전체 파일 리스트에 담기
fileList = glob.glob(filename)
```

```python
# 동일 사이즈 조절 후 이미지 데이터 읽기
dsize_ =(100,100)
imgdatas=[]

# 리사이즈, 배경제거
for file in fileList:
    try:
        org =cv2.imread(file,cv2.IMREAD_GRAYSCALE)
        img=cv2.resize(org,dsize_)
        img=remove(img)
        imgdatas.append(img)
    except Exception as e:
        print(str(e))


len(imgdatas)
```

```python
# 이미지 회전을 통해 이미지 생성
list2= []
for i in imgdatas:
    for degree in range(0,360,10):
        matrix= cv2.getRotationMatrix2D((50,50),degree,1)
        dst =cv2.warpAffine(i,matrix,(100,100))
        dst= dst/255
        dst=dst.reshape(-1)
        list2.append(dst)
```

```python
for i in range(len(list2)):
    list2[i]=np.append(list2[i],2)
```

```python
list1.extend(list2)
```

```python
import csv
with open("dino.csv","w") as f:
    writer = csv.writer(f)
    
    for i in list1:
        writer.writerow(i)
```
